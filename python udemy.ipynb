```python
#step 1: create the global variables for suits and ranks
#step 2: create the cards class which will group the suits and ranks
#step 3: create the deck class which will list out all the card combinations
#step 4: create the player class which will list out all the player card details
```

```python
#part 1
#creating the card class and initializing the global variables of the cards
import random# to shuffle the cards

suits=('Hearts','Diamonds','Spades','Clubs')
ranks=('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}
```

```python
#part 2
class Cards:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    #to return string
    def __str__(self):
        return self.rank + " of "+ self.suit
        
```

```python
#part 3
#creating a deck of cards

class Deck:
    def __init__(self):
        self.all_cards=[]#initializing a list for all cards
        
        #creating the 52 cards
        for suit in suits:
            for rank in ranks:
                created_card=Cards(suit,rank) #to get self.rank + " of "+ self.suit
                
                self.all_cards.append(created_card)
                
    #to shuffle the deck            
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    #to take one card from the deck
    #to pop an element
    def one_card(self):
        return self.all_cards.pop()
```

```python
#part 4
#creating the player class
#gives the player their list of cards
#need to pop out card and add the cards

class Player:
    
    def __init__(self,name): #name for the player's name
        self.name=name
        self.all_cards=[]
        
    def remove_card(self):
        return self.all_cards.pop(0)
    
    def add_cards(self,new_cards):
        if type(new_cards)==type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return ("{} has {} cards".format(self.name,len(self.all_cards)))
```

```python
#part 5
#Creating the basic game setup
#firstly, allot each players their cards

#creating the players
player1=Player("First person")
player2=Player("Second Person")

#creating the new deck
new_deck=Deck()
new_deck.shuffle()#contains all the cards now

#splitting the card randomly for the players

for i in range(26):
    player1.add_cards(new_deck.one_card())
    player2.add_cards(new_deck.one_card())
    
game_on=True#to start the game
```

```python
#part 6
#creating the while loop logic for the game

round=0

while game_on:#while true
    round+=1
    print("Round {}".format(round))
    
    if len(player1.all_cards)==0:
        print("{} is empty. {} wins".format(player1.name,player2.name))
        game_on=False
        break#to stop the game
    
    if len(player2.all_cards)==0:
        print("{} is empty. {} wins".format(player2.name,player1.name))
        game_on=False
        break#to stop the game
        
    #starting a card
    #the cards which the player is going to draw
    #removes a card from the deck and places it to the new list[]
    player1_drawn_cards=[]
    player1_drawn_cards.append(player1.remove_card())
    
    player2_drawn_cards=[]
    player2_drawn_cards.append(player2.remove_card())
    
    
    #to check if the card of player 1 is greater than player 2
    #vice versa or if the cards are equal
    
    at_war=True # card checking
    
    
    while at_war:
        
        #-1 is kept to take the latest card
        #0 gives the same value
        if player1_drawn_cards[-1].value>player2_drawn_cards[-1].value:
            player1.add_cards(player1_drawn_cards)
            player1.add_cards(player2_drawn_cards)
            
            at_war=False
            
        elif player2_drawn_cards[-1].value>player1_drawn_cards[-1].value:
            player2.add_cards(player2_drawn_cards)
            player2.add_cards(player1_drawn_cards)
            
            at_war=False
            
        else:
            print("Same Cards for Both, War")
            
            #can stop the game if there are less cards
            if len(player1.all_cards)<3:#giving a random number
                print("{} wins".format(player1.name))
                game_on=False
                break
            
            elif len(player2.all_cards)<3:#giving a random number
                print("{} wins".format(player2.name))
                game_on=False
                break
                
            else:#adding new cards
                for j in range(3):#random number
                    player1_drawn_cards.append(player1.remove_card())
                    player2_drawn_cards.append(player2.remove_card())
        
```

```python
#using a decorator
#decorator is used on top of another function
#the decorator is a function(already defined) 
#the decorator wraps around the current function with the old function

def fun11(func):
    print("Hello from function 1")
    
    def fun12():
        print("Inside function 1-2")
        
        func()
        
        print("Inside function 1-3")
    
    return fun12

@fun11#decorator
def fun21():
    print("Hello from function 2")
    
fun21()
```

```python
%matplotlib inline

from ipywidgets import interact, interactive
from IPython.display import clear_output, display, HTML

import numpy as np
from scipy import integrate

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import cnames
from matplotlib import animation

def solve_lorenz(N=10,angle=0.0,max_time=4.0,sigma=10.0,beta=8./3, rho=28.0):
    fig=plt.figure()
    ax=fig.add_axes([0,0,1,1],projection='3d')
    
    #preparing the axes limits
    ax.set_xlim((-25,25))
    ax.set_ylim((-35,35))
    ax.set_zlim((5,55))
    
    def lorenz_deriv(x_y_z,t0,sigma=sigma,beta=beta,rho=rho):
        
        x,y,z=x_y_z
        return[sigma*(y-x),x*(rho-z)-y,x*y-beta*z]
    
    #choosing random starting points, uniformly distributed from -15 to 15
    
    np.random.seed(1)
    x0=-15+30*np.random.random((N,3))
    
    #solve for the trajectories
    t=np.linspace(0,max_time, int(250*max_time))
    x_t=np.asarray([integrate.odeint(lorenz_deriv,x0i,t) for x0i in x0])
    
    #choose a different color for each trajectory
    
    colors=plt.cm.jet(np.linspace(0,1,N));
    
    for i in range(N):
        x,y,z=x_t[i,:,:].T
        lines=ax.plot(x,y,z,'-',c=colors[i])
        _ = plt.setp(lines,linewidth=2);
        
    ax.view_init(30,angle)
    _ = plt.show()
    
    return t,x_t
    
```

```python
t,x_t=solve_lorenz(angle=0,N=0)
```

```python
w=interactive(solve_lorenz, angle=(0.,360.),N=(0,50),sigma=(0.0,50.0), rho=(0.0,50.0))
display(w)
```
